```python
################################################################################
# Import the libraries
################################################################################

import os
import numpy as np
```

```python
################################################################################
# Load the list of Fermi IDs
################################################################################

fermi_ids = np.loadtxt('./bright_gbm_grbs.txt', dtype=str)
fermi_ids
```

```python
################################################################################
# Select the right files
################################################################################

with open('path_list.txt', 'a') as file:
    for fermi_id in fermi_ids:
        path_to_selected_units = './astrodata/romain/GBM_LC_repository/data/' + fermi_id + '/LC/selected_units.txt'
        selected_units = np.loadtxt(path_to_selected_units, dtype=str, ndmin=1)
        # print(selected_units)
        path_to_lc = './astrodata/romain/GBM_LC_repository/data/' + fermi_id + '/LC/' + fermi_id + '_LC_64ms_'
        for unit in selected_units:
            path_to_lc += unit + '_'
        path_to_lc += 'bs.txt'
        if os.path.isfile(path_to_lc):
            file.write(path_to_lc + '\n')
```
