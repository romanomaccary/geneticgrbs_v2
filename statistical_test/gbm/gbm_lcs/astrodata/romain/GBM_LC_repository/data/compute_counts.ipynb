```python
################################################################################
# Load T_start, T_stop, no. of selected units
################################################################################

t_start, t_stop = np.loadtxt('./time_list.txt', usecols=(1, 2), unpack=True)
units = np.loadtxt('./unit_list.txt', usecols=1)
```

```python
################################################################################
# Compute the counts between T_start and T_stop
################################################################################
# lcs = []
with open('./fermi_id_time_count_list.txt', 'a') as outfile:
    outfile.write('#%10s %5s %8s %8s %12s\n' % ('fermi_id', 'Units', 'T_start', 'T_stop', 'Counts'))
    for fermi_id, start, stop, unit in zip(fermi_ids, t_start, t_stop, units):
        path_to_lc = './' + fermi_id + '/LC/'
        elements = os.listdir(path_to_lc)
        elements.sort()
        # print(elements)
        if len(elements) == 2:
            lc = elements[1]
            # lcs.append(lc)
            path_to_lc += lc
            times, counts = np.loadtxt(path_to_lc, usecols=(0, 1), unpack=True)
            indices = np.where(((times >= start) & (times <= stop)))[0]
            tot_counts = np.sum(counts[indices])/unit
            print(fermi_id, tot_counts)
            outfile.write('%11s %5d %8.3f %8.3f %12.3f\n' % (fermi_id, unit, start, stop, tot_counts))
```
